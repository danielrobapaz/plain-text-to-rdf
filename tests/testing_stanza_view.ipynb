```python
# Crear el pipeline de procesamiento
nlp = stanza.Pipeline('en')

# Texto a procesar
plain_text = "The quick brown fox, which was being chased by a cunning and agile hound, jumped over the lazy dog lying near the barn."

# Medir el tiempo de procesamiento del texto
start_time = time.time()
doc = nlp(plain_text)
end_time = time.time()
print(f"Tiempo para procesar el texto: {end_time - start_time:.4f} segundos")

# Mostrar los tokens y etiquetas
print("\nTokens y Etiquetas:")

start_time = time.time()

for i, sentence in enumerate(doc.sentences):
    for token in sentence.tokens:
        print(f"Token: {token.text}")
        for word in token.words:

            # Lema de la palabra
            print(f"  Lemma: {word.lemma}")    

            # Categoria gramatical universal  
            print(f"  POS: {word.pos}")       

            # Categoria gramatical del idioma 
            print(f"  XPOS: {word.xpos}")      

            # Indice del head en el arbol
            print(f"  Head: {word.head}")       

            # Relacion de dependencia con el head
            print(f"  Deprel: {word.deprel}\n")    

end_time = time.time()
print(f"Tiempo para extraer y mostrar tokens y etiquetas: {end_time - start_time:.4f} segundos\n")

# Crear un arbol de dependencias
def to_nltk_tree(node, doc):
    if node.deprel == 'root':
        return Tree(node.text, [to_nltk_tree(child, doc) for child in doc.sentences[0].words if child.head == node.id])
    else:
        return Tree(f'{node.text} ({node.deprel})', [to_nltk_tree(child, doc) for child in doc.sentences[0].words if child.head == node.id])

# Print del arbol de dependencias (Ver en extended output)
print("Arbol de Dependencias:\n")
start_time = time.time()
for sentence in doc.sentences:
    root = [word for word in sentence.words if word.head == 0][0]
    tree = to_nltk_tree(root, doc)
    tree.pretty_print()
end_time = time.time()
print(f"\nTiempo para convertir a arbol: {end_time - start_time:.4f} segundos")
```
