```python
stanza.install_corenlp(dir=CORENLP_DIR)
stanza.download_corenlp_models(model='english-kbp', dir=CORENLP_DIR, version='main')
```

```python
plain_text = "The quick brown fox, which was being chased by a cunning and agile hound, jumped over the lazy dog lying near the barn."
```

```python
nlp = stanza.Pipeline('en')
doc = nlp(plain_text) # holds the annotation of the entire text
```

# Relation extraction

```python
from stanza.server import CoreNLPClient

text = "I miss Mox Opal"

#with CoreNLPClient(annotators=["tokenize","ssplit","pos","lemma","depparse","natlog","openie"], be_quiet=False) as client:
with CoreNLPClient(annotators=["openie"], be_quiet=False) as client:
    ann = client.annotate(text)
    #print(ann)
    for sentence in ann.sentence:
        for triple in sentence.openieTriple:
            print(triple)
```
